MEMISAHKAN DAN MENYALIN GAMBAR SESUAI DENGAN NILAI KOLOM CSV

```python
import os
import csv
import shutil

# Fungsi untuk menyalin foto ke folder sesuai dengan statusnya
def copy_images(csv_file, image_dir):
    count = 0
    with open(csv_file, 'r') as csvfile:
        reader = csv.DictReader(csvfile, delimiter=',')
        for row in reader:
            image_name = f"{row['no']}.jpeg"  # Ubah format nama gambar
            print(image_name)
            image_path = os.path.join(image_dir, image_name)
            if os.path.exists(image_path):
                status_folder = os.path.join(image_dir, row['desa'])
                os.makedirs(status_folder, exist_ok=True)
                try:
                    shutil.copy(image_path, status_folder)
                    print(f"Gambar '{image_name}' berhasil disalin ke folder '{row['desa']}'")
                    count +=1
                except shutil.Error as e:
                    print(f"Gagal menyalin gambar '{image_name}': {e}")
    return count
```

```python

if __name__ == "__main__":
    csv_file = "C:/Users/ramad/OneDrive/Dokumen/BAN_autofill/csv/oc_bondowoso.csv"
    image_dir = "C:/Users/ramad/OneDrive/Dokumen/BAN/OC Bondowoso/no tag"

    hasil = copy_images(csv_file, image_dir)
    print(hasil)
```

MEMISAHKAN DAN GAMBAR BERDASARKAN RESOLUSINYA

```python
from PIL import Image
import os

def get_image_resolution(image_path):
    with Image.open(image_path) as img:
        width, height = img.size
        return width, height

def separate_images_by_resolution(source_directory, target_directory):
    # Membuat direktori target jika belum ada
    if not os.path.exists(target_directory):
        os.makedirs(target_directory)

    # Mendapatkan daftar file gambar di direktori sumber
    image_files = [f for f in os.listdir(source_directory) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif'))]

    # Membuat dictionary untuk mengelompokkan gambar berdasarkan resolusinya
    resolution_groups = {}

    for filename in image_files:
        image_path = os.path.join(source_directory, filename)
        width, height = get_image_resolution(image_path)
        resolution = f"{width}x{height}"

        if resolution not in resolution_groups:
            resolution_groups[resolution] = []

        resolution_groups[resolution].append(filename)

    # Memindahkan gambar ke direktori tujuan berdasarkan resolusi
    for resolution, filenames in resolution_groups.items():
        resolution_directory = os.path.join(target_directory, resolution)

        # Membuat direktori resolusi jika belum ada
        if not os.path.exists(resolution_directory):
            os.makedirs(resolution_directory)

        for filename in filenames:
            source_path = os.path.join(source_directory, filename)
            destination = os.path.join(resolution_directory, filename)

            # Memindahkan gambar ke direktori tujuan
            os.rename(source_path, destination)

```

```python

if __name__ == "__main__":
    source_directory = "C:/Users/ramad/OneDrive/Dokumen/BAN/OC Bondowoso/no tag"
    target_directory = "C:/Users/ramad/OneDrive/Dokumen/BAN/OC Sumba Barat Daya/res_separate"

    separate_images_by_resolution(source_directory, target_directory)
```
