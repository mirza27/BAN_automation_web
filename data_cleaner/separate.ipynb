MEEMINDAHKAN GAMBAR BERDASARKAN FOLDER SESUAI DENGAN NILAI KOLOM CSV

```python
import os
import csv
import shutil

# Fungsi untuk menyalin foto ke folder sesuai dengan statusnya
def copy_images(csv_file, image_dir):
    count = 0
    with open(csv_file, 'r') as csvfile:
        reader = csv.DictReader(csvfile, delimiter=',')
        for row in reader:
            image_name = f"{row['No']}.jpg"  # Ubah format nama gambar
            print(image_name)
            image_path = os.path.join(image_dir, image_name)
            if os.path.exists(image_path):
                status_folder = os.path.join(image_dir, row['Desa'])
                os.makedirs(status_folder, exist_ok=True)
                try:
                    shutil.move(image_path, status_folder)
                    print(f"Gambar '{image_name}' berhasil disalin ke folder '{row['Desa']}'")
                    count +=1
                except shutil.Error as e:
                    print(f"Gagal menyalin gambar '{image_name}': {e}")
    return count
```

MEMISAHKAN DAN GAMBAR BERDASARKAN RESOLUSINYA

```python
from PIL import Image
import os

def get_image_resolution(image_path):
    with Image.open(image_path) as img:
        width, height = img.size
        return width, height

def separate_images_by_resolution(source_directory, target_directory):
    # Membuat direktori target jika belum ada
    if not os.path.exists(target_directory):
        os.makedirs(target_directory)

    # Mendapatkan daftar file gambar di direktori sumber
    image_files = [f for f in os.listdir(source_directory) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif'))]

    # Membuat dictionary untuk mengelompokkan gambar berdasarkan resolusinya
    resolution_groups = {}

    for filename in image_files:
        image_path = os.path.join(source_directory, filename)
        width, height = get_image_resolution(image_path)
        resolution = f"{width}x{height}"

        if resolution not in resolution_groups:
            resolution_groups[resolution] = []

        resolution_groups[resolution].append(filename)

    # Memindahkan gambar ke direktori tujuan berdasarkan resolusi
    for resolution, filenames in resolution_groups.items():
        resolution_directory = os.path.join(target_directory, resolution)

        # Membuat direktori resolusi jika belum ada
        if not os.path.exists(resolution_directory):
            os.makedirs(resolution_directory)

        for filename in filenames:
            source_path = os.path.join(source_directory, filename)
            destination = os.path.join(resolution_directory, filename)

            # Memindahkan gambar ke direktori tujuan
            os.rename(source_path, destination)

```

```python

if __name__ == "__main__":
    source_directory = "C:/Users/ramad/OneDrive/Dokumen/BAN/OC Bondowoso/no tag"
    target_directory = "C:/Users/ramad/OneDrive/Dokumen/BAN/OC Sumba Barat Daya/res_separate"

    separate_images_by_resolution(source_directory, target_directory)
```

```python
import os
import shutil

def pisahkan_file_berdasarkan_oc(folder_path, dest):
    # Mendapatkan daftar file di dalam folder_path
    files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

    # Memeriksa setiap file apakah mengandung kata kunci 'OC' dalam namanya
    for file_name in files:
        print(file_name)
        if 'KTP' in file_name:
            file_source_path = os.path.join(folder_path, file_name)
            file_destination_path = os.path.join(dest, file_name)

            # Memindahkan file ke dalam folder tujuan (dest)
            shutil.move(file_source_path, file_destination_path)

            print(f"File {file_name} dipindahkan ke {dest}")

# Ganti 'path_to_folder' dengan path folder yang ingin Anda proses
path_to_folder = "C:/Users/ramad/OneDrive/Dokumen/BAN/BAN NEW/tana toraja/OC"
dest_to_folder = "C:/Users/ramad/OneDrive/Dokumen/BAN/BAN NEW/tana toraja/KTP"
pisahkan_file_berdasarkan_oc(path_to_folder, dest_to_folder)
```

membuat gambar menjadi jpeg atau jpg

```python
from PIL import Image
from PIL.ExifTags import TAGS
import os

def get_geotagging_info(input_path):
    """Mengambil informasi geotag dari gambar."""
    image = Image.open(input_path)
    info = image._getexif()
    if not info:
        raise ValueError("No EXIF metadata found")

    geotagging_info = {}
    for tag, value in info.items():
        tag_name = TAGS.get(tag, tag)
        geotagging_info[tag_name] = value

    return geotagging_info

def convert_to_jpg(input_path, output_path):
    try:
        # Buka gambar menggunakan PIL
        image = Image.open(input_path)

        # Simpan sebagai file JPEG dengan menyertakan metadata
        image.convert("RGB").save(output_path, "JPEG", quality=95, exif=image.info.get("exif"))
        print(f"File {input_path} berhasil diubah menjadi {output_path}")
    except Exception as e:
        print(f"Gagal mengubah file {input_path}: {str(e)}")

def batch_convert_to_jpg(input_dir, output_dir):
    # Membuat direktori output jika belum ada
    os.makedirs(output_dir, exist_ok=True)

    # Melakukan iterasi pada setiap file di direktori input
    for filename in os.listdir(input_dir):
        input_path = os.path.join(input_dir, filename)

        # Mengecek apakah itu file dan bukan direktori
        if os.path.isfile(input_path):
            # Menghasilkan nama file output dengan ekstensi .jpg
            output_path = os.path.join(output_dir, os.path.splitext(filename)[0] + ".jpg")

            # Mengambil informasi geotag dari gambar
            # geotagging_info = get_geotagging_info(input_path)

            # Melakukan konversi ke format JPEG dengan menyertakan metadata geotag
            convert_to_jpg(input_path, output_path)

```

Mengubah jpg menjadi file

```python
from PIL import Image
from reportlab.pdfgen import canvas
import os

def convert_to_pdf(input_dir, output_dir):
    # Membuat direktori output jika belum ada
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Mengambil daftar file dalam direktori input
    files = [f for f in os.listdir(input_dir) if f.lower().endswith('.jpg')]

    for file in files:
        input_path = os.path.join(input_dir, file)
        output_path = os.path.join(output_dir, os.path.splitext(file)[0] + '.pdf')

        # Membuka gambar menggunakan Pillow
        image = Image.open(input_path)

        # Membuat file PDF menggunakan reportlab
        pdf_canvas = canvas.Canvas(output_path, pagesize=image.size)
        pdf_canvas.drawInlineImage(input_path, 0, 0)

        # Menyimpan file PDF
        pdf_canvas.save()

        print(f"{file} berhasil diubah menjadi {os.path.basename(output_path)}")
```

```python
if __name__ == "__main__":
    # Ganti direktori_input dan direktori_output sesuai dengan kebutuhan Anda
    direktori_input = "C:/Users/ramad/OneDrive/Dokumen/BAN/BAN NEW/takalar/bast"
    direktori_output = "C:/Users/ramad/OneDrive/Dokumen/BAN/BAN NEW/takalar/bast pdf version"

    convert_to_pdf(direktori_input, direktori_output)
```
